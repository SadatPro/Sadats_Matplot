```python
plt.plot(batsman['index'],batsman['RG Sharma'],label = 'RG Sharma',color ="green",marker = 'o',markersize = '12', linestyle = 'dashed',linewidth =5)
plt.plot(batsman['index'],batsman['V Kohli'],label = 'V Kohli',color ="red",marker = 'd',markersize = '12',linestyle = 'dotted',linewidth = 5)
plt.ylabel('runs')
plt.xlabel('years')
plt.title('RG Sharma vs V Kohli')
plt.legend()
plt.grid()
plt.show()
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasetml/batsman_season_record.csv')
df.head()
```

```python
plt.bar(df['batsman'],df['2015'],width = 0.5,color = 'red',label = '2015')
plt.title('batsman record')
plt.xlabel('batsman')
plt.ylabel('runs')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width = 0.2,color = 'skyblue',label = '2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2,color = 'cyan',label = '2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2,color = 'blue',label = '2017')

plt.title('batsman record')
plt.xlabel('batsman',color ='skyblue')
plt.ylabel('runs',color ='skyblue')
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width = 0.2,color = 'skyblue',label = '2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2,color = 'cyan',label = '2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2,color = 'blue',label = '2017')

plt.title('batsman record')
plt.xlabel('batsman',color ='skyblue')
plt.ylabel('runs',color ='skyblue')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width = 0.2,color = 'skyblue',label = '2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width = 0.2,color = 'cyan',label = '2016')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width = 0.2,color = 'blue',label = '2017')

plt.title('batsman record')
plt.xlabel('batsman',color ='skyblue')
plt.ylabel('runs',color ='skyblue')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
plt.bar(df['batsman'],df['2017'],color = 'skyblue',label = '2017',width= 0.5)
plt.bar(df['batsman'],df['2016'],color = 'cyan',label = '2016',width= 0.5,bottom = df['2017'])
plt.bar(df['batsman'],df['2015'],color = 'blue',label = '2015',width= 0.5, bottom = df['2016']+df['2017'])
plt.title('batsman record')
plt.xlabel('batsman',color ='skyblue')
plt.ylabel('runs',color ='skyblue')
plt.xticks(rotation = 90)
plt.legend()
# plt.show()
plt.savefig('batsman.png')
```

```python
iris = pd.read_csv('/content/drive/MyDrive/datasetml/iris.csv')
iris.sample(5)
```

```python
plt.figure(figsize=(15,8),dpi=300)
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c= iris['Species'],cmap = 'winter',alpha= 0.7)
plt.xlabel('SepalLengthCm',color = 'green')
plt.ylabel('PetalLengthCm', color = 'green')
plt.colorbar()
plt.savefig('iris.png')
```

```python
batter = pd.read_csv('/content/drive/MyDrive/datasetml/batter.csv')
```

```python
batter.shape
```

```python
sample_df = batter.head(100).sample(25,random_state= 5)
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'],c= sample_df['runs'],cmap = 'winter',alpha= 0.7)
plt.title('avg vs strike_rate')
plt.xlabel('avg')
plt.ylabel('strike_rate')
```

```python
plt.figure(figsize=(18,8),dpi=300)
plt.scatter(sample_df['avg'],sample_df['strike_rate'],c= sample_df['runs'],s= sample_df['runs'],cmap = 'winter',alpha= 0.7)
plt.title('avg vs strike_rate')
plt.xlabel('avg')
plt.ylabel('strike_rate')
plt.axhline(130, color = 'red',linestyle = '--')
plt.axhline(140, color = 'green',linestyle = '--')
plt.axvline(25, color = 'red',linestyle = '--')
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])


```

```python
fig, ax = plt.subplots(figsize=(18,8))
ax.scatter(batter['avg'],batter['strike_rate'], c ='green')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
ax.set_title('avg vs strike_rate')
fig.show()
```

```python
plt.subplots(nrows =4,ncols=4, figsize=(18,8),dpi=300)
```

```python
fig, ax = plt.subplots(nrows =2,ncols=1, figsize=(18,8),dpi=300,sharey = True)
ax[0].scatter(batter['avg'],batter['strike_rate'], c = batter['runs'],cmap = 'winter',alpha=0.5)
ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike_rate')
ax[0].set_title('avg vs strike_rate')
ax[1].scatter(batter['runs'],batter['strike_rate'], c = batter['avg'],cmap = 'winter',alpha=0.5)
ax[1].set_xlabel('runs')
ax[1].set_ylabel('strike_rate')
```

```python
fig, ax = plt.subplots(nrows =2,ncols=2, figsize=(18,8),dpi=300)
ax[0,0].scatter(batter['runs'],batter['strike_rate'])
ax[0,1].scatter(batter['runs'],batter['avg'])
ax[1,0].hist(batter['runs'])
ax[1,1].hist(batter['runs'])
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['runs'],batter['strike_rate'],cmap = 'winter',alpha=0.5)
ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batter['runs'],batter['avg'],cmap = 'winter',alpha=0.5)
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['runs'])
ax4 = fig.add_subplot(2,2,4)
ax4.hist(batter['runs'])
```
